```python
import os
import pandas as pd
from IPython.display import display

import gathering as gt
```

```python
# Mortality
mortality_df = pd.read_csv("..\data\input\child_mortality_rate.csv")
mortality_df = mortality_df.groupby(['Country', 'Year']).sum()
mortality_df = mortality_df.drop(["Unnamed: 0","Mortality Rate"], axis=1)
mortality_df["MortalityRate"] = mortality_df["Child Mortality(1 to 4)"] / mortality_df["Total Population"]
mortality_df.to_csv("..\data\output\clean_mortality.csv",index=True)
mortality_df = mortality_df.drop(['Child Mortality(1 to 4)', 'Total Population'], axis=1)
mortality_df
```

```python
# Fertility
fertility_df = pd.read_csv("..\data\input\child_natality_rate.csv")
fertility_df.set_index("Country")
fertility_df.to_csv("..\data\output\clean_fertility.csv",index=True)
fertility_df
```

```python
# Population
population_df = pd.read_csv("..\data\input\population_total_long.csv")
new_columns = {'Country Name':'Country', 'Year':'Year', 'Count':'Population'}
population_df.rename(columns=new_columns, inplace=True, errors='raise')

population_df = population_df.groupby(['Country', 'Year']).sum()
population_df.to_csv("..\data\output\clean_population.csv",index=True)

population_df
```
